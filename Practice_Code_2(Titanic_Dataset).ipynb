```python
titanic_data.fillna({'Age': titanic_data['Age'].median()}, inplace=False)
```

```python
titanic_data['Fare'].fillna(titanic_data['Fare'].median(), inplace=False)
```

```python
titanic_data = pd.read_csv('/content/tested.csv')
titanic_data = titanic_data.drop(columns=['Cabin'], errors='ignore')
titanic_data['Fare'] = titanic_data['Fare'].fillna(titanic_data['Fare'].median())
titanic_data['Sex'] = titanic_data['Sex'].astype(str).str.strip().str.lower().map({'male': 0, 'female': 1})
titanic_data['Embarked'] = titanic_data['Embarked'].astype(str).str.strip().str.upper().map({'C': 0, 'Q': 1, 'S': 2})
titanic_data['FamilySize'] = titanic_data['SibSp'] + titanic_data['Parch']
titanic_data = titanic_data.drop(columns=['PassengerId', 'Name', 'Ticket'], errors='ignore')
print(titanic_data.info())
print(titanic_data.head())
```

```python
plt.figure(figsize=(14, 10))
plt.subplot(2, 2, 1)
sns.countplot(x='Survived', data=titanic_data, palette='viridis')
plt.title('Distribution of Survival (0 = No, 1 = Yes)')
```

```python
plt.subplot(2, 2, 2)
sns.countplot(x='Pclass', data=titanic_data, palette='coolwarm')
plt.title('Distribution of Passenger Class')
```

```python
plt.subplot(2, 2, 3)
sns.countplot(x='Sex', data=titanic_data, palette='Set2')
plt.title('Distribution of Gender (0 = Male, 1 = Female)')
```

```python
plt.subplot(2, 2, 4)
sns.histplot(titanic_data['Age'], kde=True, bins=20, color='blue')
plt.title('Distribution of Age')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
plt.subplot(2, 2, 1)
sns.countplot(x='Pclass', hue='Survived', data=titanic_data, palette='Set1')
plt.title('Survival Rate by Passenger Class')
```

```python
plt.subplot(2, 2, 3)
sns.histplot(titanic_data[titanic_data['Survived'] == 1]['Age'], kde=True, color='green', label='Survived')
sns.histplot(titanic_data[titanic_data['Survived'] == 0]['Age'], kde=True, color='red', label='Not Survived')
plt.title('Survival Rate by Age')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
titanic_data['FamilySize'] = titanic_data['SibSp'] + titanic_data['Parch']
plt.figure(figsize=(6, 4))
sns.barplot(x='FamilySize', y='Survived', data=titanic_data, palette='Blues')
plt.title('Survival Rate by Family Size')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(titanic_data[titanic_data['Survived'] == 1]['Fare'], kde=True, color='green', label='Survived')
sns.histplot(titanic_data[titanic_data['Survived'] == 0]['Fare'], kde=True, color='red', label='Not Survived')
plt.title('Survival Rate by Fare')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['Fare'], kde=True, bins=30, color='orange')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='FamilySize', hue='Survived', data=titanic_data, palette='coolwarm')
plt.title('Survival Rate by Family Size')
plt.xlabel('Family Size')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(titanic_data.corr(), annot=True, cmap='RdYlGn', linewidths=0.2)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.catplot(x='Pclass', hue='Survived', col='Sex', data=titanic_data, kind='count', palette='muted')
plt.suptitle('Survival Rate by Passenger Class and Gender', y=1.03)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', hue='Survived', data=titanic_data, palette='Set2')
plt.title('Fare Distribution by Passenger Class and Survival')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=titanic_data, palette='coolwarm', alpha=0.6)
plt.title('Age vs Fare (Survival Comparison)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.pairplot(titanic_data[['Survived', 'Pclass', 'Age', 'Fare', 'FamilySize']], hue='Survived', palette='coolwarm')
plt.suptitle('Pairplot of Selected Features by Survival', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Embarked', hue='Survived', data=titanic_data, palette='Set3')
plt.title('Survival Rate by Embarked Port')
plt.xlabel('Embarked (0=Cherbourg, 1=Queenstown, 2=Southampton)')
plt.ylabel('Count')
plt.show()
```
